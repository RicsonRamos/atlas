### 🧠 1. Média Aritmética

*   **O Conceito:** É a soma de todos os valores dividida pela quantidade de observações. Representa o 'centro de massa' dos dados.
*   **O Porquê:** Usamos a média para entender o valor esperado em uma distribuição equilibrada.
*   **O Como:** Em Python, utilizamos `df['coluna'].mean()`.
*   **Relevância:** No mercado imobiliário, ela indica o volume financeiro médio, mas cuidado: ela é altamente sensível a outliers (casas de luxo), podendo distorcer a visão da 'casa comum'.

### ✂️ Seção 4 — Média Truncada (Trimmed Mean)

*   **O Conceito:** Uma variação da média que descarta uma porcentagem (ex: 10%) dos valores mais altos e mais baixos antes do cálculo.
*   **O Porquê:** Para eliminar a influência de 'ruídos' ou extremos (outliers) e focar no comportamento do mercado principal.
*   **O Como:** `scipy.stats.trim_mean(dados, 0.10)`.
*   **Relevância:** Oferece uma visão muito mais realista do mercado para um comprador comum, protegendo a análise contra a distorção causada por mansões cinematográficas.

```python
"""
Calcula a média aparada (trimmed mean) da coluna 'AdjSalePrice'.

A função remove os 10% menores e os 10% maiores valores da amostra
antes de calcular a média, reduzindo a influência de valores extremos
(outliers).
"""

from scipy.stats import trim_mean

mp = trim_mean(
    df["AdjSalePrice"],
    0.10
)
```

```python
print(f'Média: {media:,.2f}, Média Ponderada {mp:,.2f}')
```

### 🛡️ 2. Média Truncada (Trimmed Mean)

*   **O Conceito:** É a média calculada após remover uma porcentagem fixa dos menores e maiores valores do conjunto.
*   **O Porquê:** Para obter uma estimativa de tendência central que ignore os extremos sem descartar a estrutura da média.
*   **O Como:** Utilizamos `scipy.stats.trim_mean(dados, proporção)`.
*   **Relevância:** Essencial em dados com 'caudas longas' (como preços). Ela oferece uma visão mais estável do mercado ao filtrar castelos e barracos do cálculo.

### ⚖️ 3. Mediana

*   **O Conceito:** É o valor exato que divide os dados em dois grupos iguais: 50% acima e 50% abaixo.

### 🧪 Experimento: O Impacto de um Outlier Extremo

Para visualizar a robustez de cada métrica, vamos adicionar um valor hipotético de **$100.000.000** (100 milhões) ao nosso conjunto de dados e observar como as medidas se comportam.

```python
import numpy as np
from scipy.stats import trim_mean

# Criando uma série com um outlier extremo
precos_com_outlier = pd.concat([df['AdjSalePrice'], pd.Series([100000000])])

# Calculando as novas métricas
nova_media = precos_com_outlier.mean()
nova_media_truncada = trim_mean(precos_com_outlier, 0.10)
nova_mediana = precos_com_outlier.median()

print(f'--- Com Outlier de 100M ---')
print(f'Nova Média: ${nova_media:,.2f} (Antes: ${media:,.2f})')
print(f'Nova Média Truncada: ${nova_media_truncada:,.2f} (Antes: ${mp:,.2f})')
print(f'Nova Mediana: ${nova_mediana:,.2f} (Antes: ${mediana:,.2f})')
```

### 📏 4. Intervalo Interquartil (IQR)

*   **O Conceito:** É a diferença entre o terceiro quartil (75%) e o primeiro quartil (25%).
*   **O Porquê:** Medir a dispersão onde a maior parte dos dados (o 'coração' do mercado) está concentrada.
*   **O Como:** `q3 - q1` usando o método `.quantile()`.
*   **Relevância:** O IQR define a faixa de preço onde ocorre a maioria das transações 'normais'. É a base para identificar matematicamente o que é um outlier.

#### Seção 8 — MAD


```python
"""
Calcula a Mediana do Desvio Absoluto (MAD) da coluna 'AdjSalePrice'.

A função robust.scale.mad mede a dispersão dos dados de forma robusta,
ou seja, pouco sensível a outliers.

O MAD é baseado na mediana e indica o quanto os valores se afastam
tipicamente do valor central da distribuição.
"""

from statsmodels import robust

robust.scale.mad(
    df["AdjSalePrice"]
)
```

# 🚀 Nível Avançado: Inferência e Modelagem

Agora que dominamos a estatística descritiva, vamos usar esses dados para construir conhecimento preditivo.

### 🔄 16. Bootstrap: Aumentando a Confiança na Média

Na vida real, raramente temos a população inteira. O **Bootstrap** nos permite simular milhares de amostras a partir da nossa, para entender a variabilidade da nossa estimativa de média.

**O que estamos buscando?** O Intervalo de Confiança.

```python
from sklearn.utils import resample

results = []
for _ in range(1000):
    sample = resample(df['AdjSalePrice'])
    results.append(sample.median())

results = pd.Series(results)
print(f'Intervalo de Confiança (95%) para a Mediana: [${results.quantile(0.025):,.2f} - ${results.quantile(0.975):,.2f}]')

plt.figure(figsize=(10, 4))
sns.histplot(results, kde=True)
plt.title('Distribuição de Bootstrap da Mediana')
plt.show()
```

### 📈 17. Regressão Linear Múltipla

Vamos prever o `AdjSalePrice` usando as variáveis que mostraram forte correlação: `SqFtTotLiving` e `BldgGrade`.

**Equação:** $Preço = eta_0 + eta_1(Área) + eta_2(Qualidade) + eta_3(Banheiros) + eta_4(Idade)$

```python
import statsmodels.api as sm

# Preparando os dados
X = df[['SqFtTotLiving', 'BldgGrade', 'Bathrooms', 'YrBuilt']]
y = df['AdjSalePrice']

# Adicionando a constante (intercepto)
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

### 🕵️ 18. Análise de Resíduos (O Diagnóstico Final)

Um modelo avançado não é julgado pelo seu acerto (R²), mas pelos seus erros.

**Checklist de Resíduos:**
1. Os erros são aleatórios? (Se houver padrão, falta alguma variável no modelo).
2. Temos **Heterocedasticidade**? (Erros maiores para casas mais caras).

**Socrático:** Se o seu R-squared for, por exemplo, 0.55, isso significa que 55% da variação do preço é explicada pelo tamanho e qualidade. Onde estão os outros 45%? Localização? Vista para o mar?

```python
residuals = model.resid
fitted = model.fittedvalues

plt.figure(figsize=(10, 6))
sns.scatterplot(x=fitted, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Valores Previstos')
plt.ylabel('Resíduos (Erro)')
plt.title('Análise de Resíduos: O erro é constante?')
plt.show()
```

### 🚀 19. Engenharia de Variáveis: Interação

Vamos testar a hipótese de que a relação entre tamanho e preço não é apenas aditiva, mas depende da qualidade da construção.

```python
# Criando a coluna de interação
df['Living_x_Grade'] = df['SqFtTotLiving'] * df['BldgGrade']

# Preparando o novo conjunto de variáveis independentes
X_interact = df[['SqFtTotLiving', 'BldgGrade', 'Bathrooms', 'YrBuilt', 'Living_x_Grade']]
X_interact = sm.add_constant(X_interact)

# Ajustando o novo modelo
model_interact = sm.OLS(y, X_interact).fit()

print(model_interact.summary())

print(f"\nMelhoria no R-squared: {(model_interact.rsquared - model.rsquared)*100:.2f}%")
```
