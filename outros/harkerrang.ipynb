```python
import sys
import subprocess

try:
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    import time
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "selenium"])
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    import time

# Reuse the existing Chrome options if they were created earlier
if "chrome_options" not in globals():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument(
        r"--user-data-dir=C:\Users\Ramos\AppData\Local\Google\Chrome\User Data"
    )

driver = webdriver.Chrome(options=chrome_options)

try:
    # Página de inscrições
    driver.get("https://www.youtube.com/feed/channels")
    time.sleep(5)

    canais = []

    elementos = driver.find_elements(
        By.CSS_SELECTOR,
        "ytd-channel-renderer #main-link"
    )

    for e in elementos:
        nome = e.text.strip()
        url = e.get_attribute("href")
        if nome:
            canais.append(f"{nome};{url}")

    with open("inscricoes_youtube.csv", "w", encoding="utf-8") as f:
        f.write("canal;url\n")
        f.write("\n".join(canais))

    print(f"Salvos {len(canais)} canais.")

finally:
    driver.quit()
```
